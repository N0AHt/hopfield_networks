```python
import numpy as np
import matplotlib.pyplot as plt
from hopfield_net import Hopfield
```

```python
hop = Hopfield(64, random_init=True)
```

```python
print(hop.neurons.shape)
print(hop.weights.shape)
print(hop.find_energy())
```

```python
from sklearn.datasets import load_digits

digits = load_digits()
digits = (digits.images>0).astype(int)*2 - 1
print(type(digits))
print(digits.data.shape)
# plt.imshow(digits[0], cmap='gray')
# plt.show()
```

```python
patterns = digits[0:14:7]
patterns = patterns.reshape(patterns.shape[0], -1)
print(patterns.shape)

hop.update_weights_patterns(patterns)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

ax1.imshow(patterns[0].reshape(8, 8), cmap='gray')
ax1.set_title('Pattern 1')
ax1.axis('off')

ax2.imshow(patterns[1].reshape(8, 8), cmap='gray')
ax2.set_title('Pattern 2')
ax2.axis('off')

plt.tight_layout()
plt.show()
```

```python
hop.initialise_neurons()
energy = []
for i in range(4):
    energy.append(hop.find_energy())
    hop.update_neurons()
    
plt.imshow(hop.neurons.reshape(8,8), cmap = 'gray')
plt.show()

plt.plot(energy)
plt.title('Energy of the system')
plt.xlabel('Iterations')
plt.ylabel('Energy')
plt.show()

    
    
```
